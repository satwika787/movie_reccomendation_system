```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('/content/TMBD Movie Dataset.csv')
```

```python
data.head()
```

```python
require_features = data[['id', 'original_title', 'overview', 'genres']]
```

```python
require_features
```

```python
require_features['tags'] = require_features['overview']+ require_features['genres']
```

```python
require_features
```

```python
require_features = require_features.drop(columns=['overview' ,'genres'], axis=1 )
```

```python
require_features
```

```python
counter_vector = CountVectorizer(max_features= 1287 , stop_words = 'english')
```

```python
vector = counter_vector.fit_transform(require_features['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movies):
    index=require_features[require_features['original_title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:10]:
        print(require_features.iloc[i[0]].original_title)
```

```python
recommend("Avatar")
```
